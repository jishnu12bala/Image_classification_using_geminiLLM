```python
# Step 2: Import necessary modules and load YOLOv8 model
from ultralytics import YOLO
from PIL import Image

# Load the YOLOv8 pre-trained model
model = YOLO('yolov8n.pt')

# Step 3: Define a function to convert xyxy to positions
def get_position_tag(xyxy, img_width, img_height):
    x_min, y_min, x_max, y_max = xyxy
    box_center_x = (x_min + x_max) / 2
    box_center_y = (y_min + y_max) / 2
    
    # Define relative positions
    if box_center_x < img_width / 3:
        x_pos = 'left'
    elif box_center_x > 2 * img_width / 3:
        x_pos = 'right'
    else:
        x_pos = 'center'
        
    if box_center_y < img_height / 3:
        y_pos = 'top'
    elif box_center_y > 2 * img_height / 3:
        y_pos = 'bottom'
    else:
        y_pos = 'center'
    
    return f"{y_pos}-{x_pos}"

# Step 4: Classify an image and store the results
def classify_image(image_path):
    # Perform object detection
    results = model(image_path)

    
    for i, r in enumerate(results):
    # Plot results image
        im_bgr = r.plot()  # BGR-order numpy array
        im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image
        r.show()

    
    # Get the dimensions of the image
    img = results[0].orig_img
    img_height, img_width, _ = img.shape
    
    # Arrays to store class names and positions
    class_names = []
    positions = []
    
    # Loop over each detection
    for detection in results[0].boxes:
        # Extract class ID, confidence, and bounding box (xyxy)
        class_id = int(detection.cls)
        xyxy = detection.xyxy[0].numpy()  # Bounding box in xyxy format
        
        # Get the class name and position tag
        class_name = model.names[class_id]
        position_tag = get_position_tag(xyxy, img_width, img_height)
        
        # Store in arrays
        class_names.append(class_name)
        positions.append(position_tag)
    
    return class_names, positions

# Step 5: Test the function with an image
image_path = 'img.jpg'  # Replace with the path to your image
class_names, positions = classify_image(image_path)

# Output the results
print("Class Names:", class_names)
print("Positions:", positions)


```
